```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.ylim(0, 120)
plt.xlim(0, 7)
```

```python
np.random.seed(4)
# x and y are some random points to get a polynomial curve.
x = np.random.normal(3.0, 1.0, 1000)
y = np.random.normal(50.0, 10.0, 1000) / x
a = np.array(x)
b = np.array(y)
```

```python
p9 = np.poly1d(np.polyfit(a, b, 9)) # Polyfit takes degree as it's 3rd Parameter.
xp = np.linspace(0, 6, 300)
```

```python
plt.scatter(x, y, s=15)
plt.plot(xp, p9(xp), color='red')
```
